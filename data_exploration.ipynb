### Distribution of channel
The distribution of channels is also not very balanced. There are some that are very common, up to a few 1000 times and others that appear only once.

```python
plt.figure(figsize=(20,3))
plt.subplot(211)
dat['channel'].value_counts()[:80].plot('bar')
plt.subplot(212)
dat['channel'].value_counts()[80:].plot('bar')
plt.show()
```

## Converting dates

```python
dat['weekday'] = dat['click_time'].dt.weekday.astype('category', categories=range(7), ordered=True)
dat['day'] = dat['click_time'].dt.day
dat['hour'] = dat['click_time'].dt.hour
```

```python
dat[dat['is_attributed'] == 1]['weekday'].value_counts().plot('bar')
plt.show()
```

```python
dat[dat['is_attributed'] == 0]['weekday'].value_counts().plot('bar')
plt.show()
```
